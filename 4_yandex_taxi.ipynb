Антон пользуется тарифом UberX в Uber.Russia, а Андрей — тарифом Эконом в Яндекс.Go. Проверим, как параметры тарифа влияют на цену.

Уточнение: не вполне понятно из задания является ли дистанция 16.5 км. от дома до офиса прямой или это дистанция маршрута на автомобиле. Поэтому приведем расчет обоих вариантов.

```python
def count_tariff(distance, time):
  print(f'Yandex:  {int((distance-3)*8) + int((time-6)*10) + 139}')
  print(f'Uber:   {int((distance-4)*12) + int((time-5)*6) + 129}')
  print(f'Время в дороге: {int(time)} мин.')
  print(f'Дистанция: {round(distance, 2)} км.')
  print(f'Разница в рублях: {(int((distance-3)*8) + int((time-6)*10) + 139) - \
                         (int((distance-4)*12) + int((time-5)*6) + 129)}')
  print('\n')
  print(10 * "*")
  print('\n')
```

```python
# Предположим, что 16.5 км. это дистанция по прямой и длина маршрута на такси
# может отличаться
for a in np.linspace(20, 40, 5):
  for b in np.linspace(17, 25, 4):
    count_tariff(b,a)
```

```python
# Предположим, что 16.5 км это длина маршрута по дороге
for a in np.linspace(15,50,10):
  count_tariff(16.5,a)
```

Таким образом, у Андрея, который пользуется такси Yandex Go, цена за поездку будет выше практически всегда. Тарификация данных тарифов устроена таким образом, что Yandex Go оказывается дешевле только на средних расстояниях с небольшим временем в пути.
